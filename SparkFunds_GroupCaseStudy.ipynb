```python
companies_columns_to_remove_special_char = ['permalink','name']
rounds2_columns_to_remove_special_char = ['company_permalink']

companies[companies_columns_to_remove_special_char] = companies[companies_columns_to_remove_special_char].applymap(lambda x 
                                                                                               : str(x).encode('utf-8').decode('ascii','ignore'))
rounds2[rounds2_columns_to_remove_special_char] = rounds2[rounds2_columns_to_remove_special_char].applymap(lambda x 
                                                                                               : str(x).encode('utf-8').decode('ascii','ignore'))
```

```python
# companies df duplicate count case-insensitive

companies_columnss_to_check_duplicates = ['permalink','name','homepage_url']
pd.DataFrame(list(map(lambda x : 
                      (x, np.sum(companies[x].str.upper().duplicated())), 
                      companies_columnss_to_check_duplicates)), 
             columns=['col_name', 'duplicate_count'])
```
